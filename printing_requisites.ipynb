```python
import requests
import json
```

```python
api_url = "https://handbook.monash.edu/api/es/search"
```

```python
request_json = {
    "query": {
        "bool": {
            "must": [
                {"term": {"live": True}},
                [
                    {
                        "bool": {
                            "minimum_should_match": "100%",
                            "should": [
                                {
                                    "query_string": {
                                        "fields": [
                                            "monash2_psubject.implementationYear"
                                        ],
                                        "query": "*2022*",
                                    }
                                }
                            ],
                        }
                    }
                ],
            ],
            "filter": [{"terms": {"contenttype": ["monash2_psubject"]}}],
        }
    },
    "sort": [{"monash2_psubject.code_dotraw": {"order": "asc"}}],
    "from": 0,
    "size": 50, # change this line to however many units you want to get back in 1 call
    "track_scores": True,
    "_source": {
        "includes": [
            "*.code",
            "*.name",
            "*.award_titles",
            "*.keywords",
            "urlmap",
            "contenttype",
        ],
        "excludes": ["", None],
    },
}
```

```python
response = requests.post(api_url, json=request_json)
```

```python
response_json = response.json()
```

```python
for unit in response_json["contentlets"]:
    unit_code = unit["code"]
    unit_data_str = unit["data"]
    unit_data_dict = json.loads(unit_data_str)
    print("Unit: ", unit_code)
    
    for requisite in unit_data_dict["requisites"]: 
        requisite_name = requisite["requisite_type"]["label"]
        print(requisite_name)
        
        for item in requisite["container"]:
            connector = " " + item["parent_connector"]["label"] + " "
            related_units = []
            for unit in item["relationships"]:
                related_units.append(unit["academic_item_code"])
            output = connector.join(related_units)
            print(output)
                
            
```
